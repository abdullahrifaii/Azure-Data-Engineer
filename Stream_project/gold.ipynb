```python
gold_df = spark.sql("""
SELECT
    city,
    country,
    date(weather_time) AS weather_date,
    COUNT(*) AS readings_count,
    ROUND(AVG(temperature_c), 2) AS avg_temperature_c,
    ROUND(MAX(temperature_c), 2) AS max_temperature_c,
    ROUND(MIN(temperature_c), 2) AS min_temperature_c,
    ROUND(AVG(humidity_pct), 2) AS avg_humidity_pct,
    ROUND(AVG(wind_speed_kmh), 2) AS avg_wind_speed_kmh,
    SUM(CASE WHEN is_rainy THEN 1 ELSE 0 END) AS rainy_readings,
    SUM(CASE WHEN is_windy THEN 1 ELSE 0 END) AS windy_readings
FROM weather_silver
GROUP BY city, country, date(weather_time)
""")

gold_df.write.mode("overwrite").format("delta").saveAsTable("meteo.gold.weather_gold")
```
