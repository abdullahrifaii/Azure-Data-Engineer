```python

nested_list = [[1,2,3,4],[2,7,6],[1,2]]
```

```python
flattened_list = [i for sublist in nested_list for i in sublist]
flattened_list
```

```python
len_of_nes = [len(i) for i in nested_list]
len_of_nes
```

```python
dict_len = {i:len(sublist) for i, sublist in enumerate(nested_list)}
dict_len
```

# Get the common elements of two lists

```python
list1 = [1,2,3,3,4,5,1]
list2 = [9,7,8,5,4,1]

common = list(set(list1) & set(list2))
common
```

```python
commlis = [i for i in list1 if i in list2]
commlis
```

```python
list_rep = [1,1,1,2,3,3,2,4,5,5,7,7,7,7,7,9]

dict1 = {i:list_rep.count(i) for i in list_rep}
dict1
```

```python
empdic = {}
for item in list_rep:
  empdic[item] = empdic.get(item,0) + 1
print(empdic)
```
